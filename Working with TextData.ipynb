<h3>More Practice with splits() </h3>

```python
import pandas as pd
chicago = pd.read_csv("chicago.csv").dropna(how = "all")
chicago['Department'] = chicago['Department'].astype('category')
chicago.tail()
```

```python
#Extracting and counting the Lastname from the 'Name' column
chicago['Name'].str.split(",").str.get(0).value_counts().head(3)
```

```python
#We want to get the Firstname from the series and get count of the name which repeated most of the times
chicago['Name'].str.split(",").str.get(1).str.strip().str.split(" ").str.get(0).value_counts().head(3)
```

<h3>The expand and n parameters of the str.split() Method. </h3>

```python
import pandas as pd
chicago = pd.read_csv("chicago.csv").dropna(how = "all")
chicago['Department'] = chicago['Department'].astype('category')
chicago.tail()
```

```python
#The expand parameter of split() method , returns a dataframe rather than series or python list
chicago[["FirstName","lastName"]] = chicago['Name'].str.split(",",expand = True)
```

```python
chicago.head(3)
```

```python
#Since the Position title column has more than one word as values and also it has multiple spaces after the word
#the split method creates multiple columns in order to encorporate spaces
chicago['Position Title'].str.split(" ",expand = True)
```

```python
#The 'n' inclues the max number of splits we want to take for a column values
chicago['Position Title'].str.split(" ",expand = True, n = 1)
```

```python
#The 'n' inclues the max number of splits we want to take for a column values
chicago[["FirstTitleWord","Remaining word"]] = chicago['Position Title'].str.split(" ",expand = True, n = 1)
```
