<h3>5.Parameters and Arguments </h3>

```python
fruits = ['Apple','Banana','Kiwi','Orange','Pineapple','Mango']
weekdays = ['Sunday','Monday','Tuesday','wednesday','Thrusday','Friday']
#Here As per the series constructor , the first argument is taken as data and second argument is taken as index
#If the second argument is less than or greater than the data values ERROR is thrown
pd.Series(fruits,weekdays)
```

```python
fruits = ['Apple','Banana','Kiwi','Orange','Pineapple','Mango']
weekdays = ['Sunday','Monday','Tuesday','wednesday','Thrusday','Friday']
pd.Series(data = fruits,index = weekdays)
```

```python
fruits = ['Apple','Banana','Kiwi','Orange','Pineapple','Mango','Grape']
weekdays = ['Sunday','Monday','Tuesday','wednesday','Thrusday','Friday','Monday']
# Here we have duplicate index , Some operations are restricted if we have duplicate index
pd.Series(data = fruits,index = weekdays)
```

<h3>16.The .get() Method on a series </h3>

```python
Pokemon = pd.read_csv("pokemon.csv",index_col = 'Pokemon',squeeze = True)
Pokemon.sort_values()
Pokemon.head(3)
```

```python
Pokemon.get('Venusaur')
```

```python
Pokemon.get(['Venusaur','Ivysaur'])
Pokemon.get(key = ['Venusaur','Ivysaur'])
#Both the methods return same result
```

```python
#The get() Method doesnt return an error , if we are trying to retrieve a value that doesnt exists
#Since the default value is set as 'None'
Pokemon.get('Digimon')
```

```python
Pokemon.get('Digimon',default  = "This is not a Pokemon value")
```

```python
#If we pass a list for which few values exist and other values doesnt exist , it returns NaNfor the values that doesnt exist
Pokemon.get(key = ['Ivysaur','Digimon'],default  = "This is not a Pokemon value")
```

<h3>20. The .apply() method

```python
google = pd.read_csv("google_stock_price.csv",squeeze = True)
google.head(3)
```

```python
def classify_performance(number):
    if number < 300:
        return 'Ok'
    elif number > 350 and number < 650:
        return "Satisfactory"
    else:
        return "Incredible"
```

```python
google.apply(classify_performance).tail()
```

```python
google.head(4)
```

```python
google.apply(lambda stock_price : stock_price + 1) #Anomynous Function
```

<h3>21. The map() method </h3>

```python
pokemon_names = pd.read_csv("pokemon.csv",usecols = ['Pokemon'],squeeze = True)
pokemon_names.head()
```

```python
pokemon_types = pd.read_csv("pokemon.csv",index_col = 'Pokemon',squeeze = True)
pokemon_types.head()
```

```python
pokemon_names.map(pokemon_types)
```

```python
#Working with dictionary mapping
pokemon_names = pd.read_csv("pokemon.csv",usecols = ['Pokemon'],squeeze = True)
pokemon_types = pd.read_csv("pokemon.csv",index_col = 'Pokemon',squeeze = True).to_dict() 
#converting a series to dictionary
```

```python
pokemon_names.head()
```

```python
pokemon_types #Returns python dictionary
```

```python
pokemon_names.map(pokemon_types) #map method maps the series to any data collection
```
